## Q.4

-Plot Stacked Bar Graph for the above data.
    - Use proper colors and x and y label for the graph
    - Use legend positioned at upper right corner
    - Also mark xticks with an angle of 90 Degree

```python
# Answer 4

labels = ['Delhi','Mumbai','Chennai','Kolkata','Hyderabad']
attr = ['Clothing','Equipment','Accessories']
Clothing = np.array([82,78,48,46,39])
Equipment = np.array([48,31,51,29,18])
Accessories = np.array([15,20,14,13,9])
ind = [a for a,_ in enumerate(labels)]
plt.bar(ind, Clothing, width=0.5, label='Clothing', color='b', bottom=Equipment+Accessories)
plt.bar(ind, Equipment, width=0.5, label='Equipment', color='y', bottom=Accessories)
plt.bar(ind, Accessories, width=0.5, label='Accessories', color='k')
plt.xlabel('Cities')
plt.xticks(ind,labels,rotation='90')
plt.ylabel('Stores Range')
plt.legend(attr,title="Stores")
plt.title('Stacked bar graph')
plt.show()
```

## Q.5 For Titanic Dataset
- Load Dataset using Seaborn
- Plot a Bar Graph of Survival of Passenger against the Passenger Class.
- Use proper "hue" attribute, and palette color. Mark 'x' and 'y' label with proper title.
- Use legend located at upper left corner

```python
# Answer 5

# Loading the Dataset - provided by seaborn
data=ss.load_dataset('titanic')

Pclass=list(data.pclass.unique())
Pclass.sort()
# Distributing the dataset according to the Pclass & the Survival
Survival0=[data[(data['survived']==0) & (data['pclass']==1)].pclass.count(),data[(data['survived']==0) & (data['pclass']==2)].pclass.count(),data[(data['survived']==0) & (data['pclass']==3)].pclass.count()]
Survival1=[data[(data['survived']==1) & (data['pclass']==1)].pclass.count(),data[(data['survived']==1) & (data['pclass']==2)].pclass.count(),data[(data['survived']==1) & (data['pclass']==3)].pclass.count()]

width=0.40
x = np.arange(len(Pclass))
plt.bar(x-.2,Survival0,width,label='NO',color='b')
plt.bar(x+.2,Survival1,width,label='YES',color='lightblue')
# graph will contain the grid which increses readability
plt.grid(True)
plt.legend(title='Survival',loc='upper left')
plt.xticks(x,Pclass)
plt.xlabel("Class Of Passenger")
plt.ylabel("No. of passengers")
plt.title("Bar Graph showing the Survival of Passenger against the Class of the Passengers",color='b',size=15)
plt.show()
```

## Q.6 For Tips Dataset
- Load the Dataset Using Seaborn
- Plot the Horizontal Distribution Graph on Total Bill with color as Green
- Mark proper 'x' and 'y' label and give a suitable title

```python
# Answer 6

# Loadin the dataset
tip=ss.load_dataset('tips')

ss.distplot(tip['total_bill'],color='g',vertical=True)
plt.title('Distribution Of the Total Bill',color='g',size=20)
plt.ylabel('Amount Of Bill',color='b',size=15)
plt.xlabel('Mean Of Bill',color='b',size=15)
plt.show()
```

## Q.9 


- Plot the Pie Chart for the two colums with some explode for the country "India"  and proper shadow

```python
# Answer 9

countries = ['United States','Great Britain','India','Russia','Germany']
medals = [46,27,26,19,17]
explode = [0,0,0.2,0,0]
colors = ['red','yellow','blue','skyblue','lightgreen']
plt.pie(medals,colors=colors,autopct='%1.1f%%',pctdistance=0.7,shadow=True,explode=explode,textprops={'fontsize': 15},wedgeprops={"edgecolor":"0",'linewidth': 1,'linestyle': 'solid', 'antialiased': True})
plt.axis('equal')
plt.legend(countries,title="Countries",loc='upper right')
plt.title('Percentage of Medals won by Countries',color='b',size=15)
plt.show()
```

## Q.10 For Tips Dataset
- Load the dataset using seaborn
- Draw the Strip Plot between 'total bill' and 'day'
- Rotate xticks with an angle of 45 Degree

```python
# Answer 10

# Loading the DataSet
tips = ss.load_dataset('tips')

# Plotting the StripPlot
ss.stripplot(x='day',y='total_bill',data=tips)
plt.xlabel('Day Of the Week',color='b',size=15)
plt.ylabel('Total Bill Amount',color='b',size=15)
#rotating the x axis pointer by 45 degree to increase readability
plt.xticks(rotation=45)
plt.show()
```
