```python
api.drop(['City','CityCode','Lat','Lon','Province','CountryCode'],axis=1,inplace=True)
api.tail(2)
```

```python
def normalz(obj):
    obj = list(obj)
    for i in range(1,len(obj)-1):
        x,y,z = obj[i-1],obj[i],obj[i+1]
        if (x<y and y>z and z>x) or (x>y and y<z and z>x) :
            obj[i] = (x+z)//2
    return pd.Series(obj)

api['Confirmed'] = normalz(api['Confirmed'])
api['Active'] = normalz(api['Active'])
api['Recovered'] = normalz(api['Recovered'])
api['Deaths'] = normalz(api['Deaths'])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
```

```python
api.head(2)
```

## Linear Regression

```python
y = api.iloc[:,1:2].values
x = api.iloc[:,6:].values
plt.scatter(x,y)
plt.show()
```

```python
reg = LinearRegression()
reg.fit(x,y)
ypred = reg.predict(x)
```

```python
plt.scatter(x,y)
plt.plot(x,ypred, color='r')
plt.show()
```

```python
r2_score(y,ypred)*100
```

* As by above plot we can see that Linear Regression is not a good way to predict the curve as its accuracy is really very low i.e. 63%

## Let's try Polynomial Regression

```python
# splitting the data into train/test
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)

# making instance of PloyNomial Regression
poly_features = PolynomialFeatures(degree=4) # change degree here to try different regression lines

# fitting the data in regressor
x_poly = poly_features.fit_transform(x_train)

reg1 = LinearRegression()
reg1.fit(x_poly,y_train)
```

```python
plt.scatter(x_train, y_train, color='r')
plt.plot(x_train, reg1.predict(x_poly), color='blue')
plt.xlabel("Days")
plt.ylabel("Confirmed Cases India")
plt.show()
```

```python
# taking values predicted by the model
y_pred = reg1.predict(poly_features.fit_transform(x_test))

# checking the accuracy of model
r2_score(y_test,y_pred)*100
```

* degree=2 accuracy = 93.08%
* degree=3 accuracy = 99.72%
* degree=4 accuracy = 99.98%
* degree=5 accuracy = 99.97%
* degree=6 accuracy = 99.97%
* degree=7 accuracy = 99.98%
* degree=8 accuracy = 99.98%
* degree=9 accuracy = 99.98%

As by above data, we see that degree=4 & above are giving really good accuracy, so we'll use ploynomial regression with degree 4

### Let's predict confimed cases curve for future 1 year

```python
x_future_df = pd.DataFrame({"Days":np.arange(1,366)})
x_future_df.head()
```

```python
x_future = x_future_df.iloc[:,0:].values
y_future = reg1.predict(poly_features.fit_transform(x_future))

plt.plot(x_future,y_future, color='r')
plt.xlabel("Days",color='b',size=15)
plt.ylabel("Confirmed Cases: in Crores",color='b',size=15)
plt.title("Predicted Confirmed Cases in India till 22/01/2021",color='r',size=20)
plt.show()
```

```python
x_future_df['Predicted_Confirmed'] = y_future
x_future_df.tail()
```

* So, according to the Analysis, in India if no cure is found for Covid-19 then according to present rate, there will be around 2.96 Crore Confirmed Cases on 22 Jan 2021
