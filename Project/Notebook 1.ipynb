## Data Wrangling

```python
states = states[states.State != 'Total']
```

```python
agrp.set_index('Sno',inplace=True)
agrp.drop('TotalCases', axis=1, inplace=True)
agrp = agrp.replace({'Percentage': '[A-Za-z %]' },"", regex=True)
agrp.head(2)
```

```python
agrp = agrp.replace({'AgeGroup':'Missing'},"Super Senior Citizen")
agrp.tail(2)
```

```python
zone.drop(['districtcode','source'],axis=1,inplace=True)
zone.head(2)
```

```python
def clear_time(s):
    s = s.split(" ")
    s = s[0]
    return s

country['Last Update'] = country['Last Update'].apply(clear_time)
country.head(2)
```

```python
country.drop(['ObservationDate'], axis=1, inplace=True)
country.head(2)
```

```python
def float_to_int(s):
    return int(s)
country['Confirmed'] = country['Confirmed'].apply(float_to_int)
country['Deaths'] = country['Deaths'].apply(float_to_int)
country['Recovered'] = country['Recovered'].apply(float_to_int)
country.head(2)
```

```python
country.tail(5)
```

```python
l = list(beds.columns)
l.remove('TotalPublicHealthFacilities_HMIS')
l.remove('NumPublicBeds_HMIS')
l.remove('State/UT')
beds.drop(l,axis=1,inplace=True)
beds.rename(columns={'TotalPublicHealthFacilities_HMIS':'Total Health Centers', 'NumPublicBeds_HMIS': 'Total Beds'},inplace=True)
beds.head(5)
```

# Data Visualization

```python
plt.figure(figsize = (10,10))
wedgeprops = {
    "linewidth": 3,
    'width':1550,
    "edgecolor":"k"
}

plt.pie(
    agrp['Percentage'],
    shadow=True,
    autopct='%1.1f%%',
    radius=1800,
    labels=agrp['AgeGroup'],
    wedgeprops = wedgeprops,
    startangle=-10,
    rotatelabels=True,
    pctdistance = 0.7,
)
plt.legend(agrp['AgeGroup'], title="Age Groups (years)", prop={"size":8})
plt.axis('equal')
plt.title("Age Groups Suffering from Covid-19 India",color='k',size=15)
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (16, 4)

plt.plot(api['Date'],api['Confirmed'], color='r')
plt.plot(api['Date'],api['Active'], color='y')
plt.plot(api['Date'],api['Recovered'], color='g')
plt.plot(api['Date'],api['Deaths'], color='k')

plt.xlabel("Dates",size=15,color='b')
plt.ylabel("No. of Cases",size=15,color='b')
plt.title("India Covid-19 Cases till Date",size=20)
plt.grid()
plt.legend(title='Cases')
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (16, 4)

plt.bar(states['State'],states['Confirmed'],color='r')

plt.xlabel('States',size=18)
plt.ylabel('Total Confirmed Cases', size=18)
plt.xticks(rotation=90)

plt.show()
```
